```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
data = pd.read_csv('Student_Performance.csv')
```

```python
df = pd.DataFrame(data)
df
```

```python
x = df.drop(['Performance Index', 'Extracurricular Activities'], axis = 1)
y = df['Performance Index']
```

```python
df['Extracurricular Activities'] = df['Extracurricular Activities'].map({
    'Yes': 1,
    'No': 0
})
df
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.25, random_state=42 )
```

```python
x_train.shape , x_test.shape, y_train.shape, y_test.shape
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
r2_score(y_test, y_pred)*100
```

```python
mean_squared_error(y_test, y_pred)
```

```python
# r = ((y_act - y_pred)**2).mean()
```

```python
import numpy as np
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
data = pd.read_csv('titanic.csv')
```

```python
df1 = pd.DataFrame(data)
df1.head()
```

```python
df1.info()
```

```python
df1.duplicated().sum()
```

```python
df1.isna().sum()/df1.shape[0]
```

```python
df1.drop("Cabin",axis=1 ,inplace = True)
```

```python
df1['Age'].mean()
```

```python
df1.fillna({'Age':29}, inplace= True)
```

```python
df1['Age'].fillna(29, inplace =True)
```

```python
df1
```

```python
df1.fillna({'Embarked': df1['Embarked'].mode()[0]}, inplace= True)
```

```python
df1.drop(['PassengerId','Name'],axis=1, inplace=True)
```

```python
df1['Sex'] = df1['Sex'].astype(str).str.lower().str.strip()

df1['Sex'] = df1['Sex'].map({
    'male': 1,
    'female': 0
})

```

```python
df1
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
le = LabelEncoder()
df1['Embarked']=le.fit_transform(df1['Embarked'])
```

```python
df1
```

```python
df1.drop(['Ticket'],axis=1, inplace=True)
```

```python
df1
```

```python
y = df1['Survived']
x = df1.drop(['Survived'],axis=1)

df1
```

```python
scale = StandardScaler()
```

```python
x_scaled = scale.fit_transform(x)
x_scaled = pd.DataFrame(x_scaled, columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
precision_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```
